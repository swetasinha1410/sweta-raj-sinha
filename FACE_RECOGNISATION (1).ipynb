OpenCV provides pre-trained models on Haar features and Cascade classifiers.

**2. Load Libraries**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**3. Use Cascadeclassifier**

```python
face_cascade = cv2.CascadeClassifier("/content/haarcascade_frontalcatface_extended.xml")
```

**4. Load an image using openCV**

```python
image = cv2.imread("/content/cat.jpg")
```

```python
gray_img = cv2.cvtColor(image ,cv2.COLOR_BGR2GRAY) #convert an image from one color space to another.
```

**5. Apply face detection techniqe on loaded image**

```python
faces = face_cascade.detectMultiScale(gray_img, scaleFactor = 1.05, minNeighbors = 5)
```

gray is the input grayscale image.

scaleFactor is the parameter specifying how much the image size is reduced at each image scale. It is used to create the scale pyramid.

minNeighbors is a parameter specifying how many neighbors each candidate rectangle should have, to retain it. A higher number gives lower false positives.

```python
if len(faces) == 0:
    print("No faces found")
 
else:
    print("Number of faces detected: " + str(faces.shape[0]))

    # blue rectangle “mark” the faces in the image
    
 
    for (x, y, w, h) in faces: 
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0),3)
        #color: It is the color of border line of rectangle to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for blue color.
        #thickness=3
        
        plt.imshow(image)
```

If faces are found, it returns the positions of detected faces as a rectangle with the left up corner (x,y) and having “w” as its Width and “h” as its Height ==> (x,y,w,h). Please see the picture above.
