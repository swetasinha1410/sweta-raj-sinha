**Import Libraries**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2 # for image to load
```

**Load Image**

```python
img = cv2.imread("/content/dog.PNG")
plt.imshow(img)
```

```python
img.shape # checking size of image
```

**Preprocessing**

```python
data=img/255
# divide all pixel value by 255, 255 will be maximum,so that pixel value normalised between 0 & 1
# when their is absolute  white space it will transform 1
# when their is absolute  black space it will transform 0
# rest will be in between 0 to 1
```

```python
data=data.reshape(490*899,3) 
```

```python
data.shape
```

Here after re-shape 440510 are row and 3 are columns . Now it becomes regular structure data , as image is a generall not structure data.
And now we can applly K-Means cluster Algo.

**K-means cluster**

```python
KMeans=KMeans(6)
KMeans.fit(data)
new_colors=KMeans.cluster_centers_[KMeans.predict(data)]
```

```python
new_colors.shape
```

```python
dog_recolor=new_colors.reshape(img.shape) # to plot a image in original shape
```

```python
dog_recolor.shape
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,6),subplot_kw=dict(xticks=[], yticks=[]))

fig.subplots_adjust(wspace=0.05)
ax[0].imshow(img)
ax[0].set_title("Original",size=16)
ax[1].imshow(dog_recolor)
ax[1].set_title("3D-coloured image",size=16)
```

**Observation : More the number of clusters, better the clarity of the image 😊**

**Checking with less Number of cluster🤞**

```python
KMeans2=KMeans(4)
KMeans2.fit(data)
new_colors2=KMeans2.cluster_centers_[KMeans2.predict(data)]
```

```python
new_colors2.shape
```

```python
dog_recolor2=new_colors2.reshape(img.shape) # to plot a image in original shape
dog_recolor2.shape
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,6),subplot_kw=dict(xticks=[], yticks=[]))

fig.subplots_adjust(wspace=0.05)
ax[0].imshow(img)
ax[0].set_title("Original",size=16)
ax[1].imshow(dog_recolor2)
ax[1].set_title("3D-coloured image",size=16)
```

**Observation-Image segmentation with less  are not  clear as in above , more number of clusters**
