**KDE Plot Visualization with Pandas and Seaborn**

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

Here i will be using Iris Dataset and KDE Plot to visualize the insights of the dataset.

About the Iris Dataset –

Attributes : Petal_Length (cm), Petal_Width (cm), Sepal_Length (cm), Sepal_Width(cm)
Target : Iris_Virginica, Iris_Setosa, Iris_Vercicolor
Number of Instances : 150
One-Dimensional KDE Plot :

**One-Dimensional KDE Plot :**

We can visualize the probability distribution of a sample against a single continuous attribute.

```python
# importing the required libraries
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Setting up the Data Frame
iris = datasets.load_iris()
  
iris_df = pd.DataFrame(iris.data, columns=['Sepal_Length',
                      'Sepal_Width', 'Patal_Length', 'Petal_Width'])

iris_df
```

```python
iris_df['Target'] = iris.target
  

iris_df
  
```

```python
iris_df['Target'].replace([0], 'Iris_Setosa', inplace=True)
iris_df['Target'].replace([1], 'Iris_Vercicolor', inplace=True)
iris_df['Target'].replace([2], 'Iris_Virginica', inplace=True)
iris_df
```

```python
iris_df.columns
```

```python
# Plotting the KDE Plot
plt.figure(figsize = (4,4))
sns.kdeplot(iris_df.loc[(iris_df['Target']=='Iris_Virginica'),'Sepal_Length'], color='g', shade=True, Label='Iris_Virginica')

  
# Setting the X and Y Label
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```

```python
iris_df1=pd.DataFrame(iris_df,columns=['Sepal_Length', 'Sepal_Width', 'Patal_Length', 'Petal_Width', 'Target'])
iris_df1

```

```python
iris_df1['Target'].unique()
```

```python
Iris_Setosa = iris_df1[iris_df1["Target"] == "Iris_Setosa"]
Iris_Vercicolor = iris_df1[iris_df1["Target"] == "Iris_Vercicolor"]
Iris_Virginica = iris_df1[iris_df1["Target"] == "Iris_Virginica"]

plt.figure(figsize = (12,12))
for ax, col in enumerate(iris_df.columns[:4]):
    plt.subplot(4,4, ax + 1)# here 4,4 will result 4 graphs 
    plt.title(col)
    sns.kdeplot(x = Iris_Virginica[col], label = "Iris_Virginica")
    sns.kdeplot(x = Iris_Vercicolor[col], label = "Vercicolor")
    sns.kdeplot(x = Iris_Setosa[col], label = "Iris_Setosa")

    plt.legend()
plt.tight_layout()
```
