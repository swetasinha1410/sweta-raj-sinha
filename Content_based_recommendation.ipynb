```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df=pd.read_csv("/content/sample-data.csv")
```

```python
df.head()
```

```python
vectorizer=TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix=vectorizer.fit_transform(df['description'])
```

```python
tfidf_matrix
```

500 row and 52262 is unique tokens

```python
cosine_similarity=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
results={}
for idx , row in df.iterrows():
  similar_indices=cosine_similarity[idx].argsort()[:-100:-1]
  simailar_items=[(cosine_similarity[idx][i],df['id'][i]) for i  in similar_indices] 
  results[row['id']] = simailar_items[1:]
```

```python
results
```

```python
def item(id):
    return df.loc[df['id'] == id]['description'].tolist()[0].split(' - ')[0]
```

```python
# Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Recommending " + str(num) + " products similar to " + item(item_id) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: " + item(rec[1]) + " (score:" + str(rec[0]) + ")")
```

```python
recommend(item_id=5, num=10)
```

```python
recommend(item_id=9, num=10)
```
